```python
numerical_feature = [feature for feature in df.columns if df[feature].dtypes!='O']
numerical_feature
```

```python
categorical_features = [feature for feature in df.columns if df[feature].dtypes=='O']
categorical_features
```

## one hot encoding

```python
df1 = df[categorical_features].iloc[:,1:7]
df1
```

```python
data = df1.copy()
```

```python
for col in df1.columns:
    print(f"{col}: {len(df1[col].unique())} labels")
```

```python
## let's get how many columns we will get after one hot encoding
pd.get_dummies(df1,drop_first=True).shape
```

```python
df1.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
top_10 = [x for x in df1.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
for label in top_10:
    df1[label] = np.where(df1['X2']==label,1,0)
df1[['X2']+top_10].head(20)
```

```python
def one_hot_top_x(df,variable,top_x_labels):
    for label in top_x_labels:
        df[variable+'_'+label] = np.where(df[variable]==label,1,0)
data = df1.copy()
one_hot_top_x(data,'X2',top_10)
data.head()
```

```python
## for X1
top_10 = [x for x in df1.X1.value_counts().sort_values(ascending=False).head(10).index]
one_hot_top_x(data,'X1',top_10)
data.head()
```
