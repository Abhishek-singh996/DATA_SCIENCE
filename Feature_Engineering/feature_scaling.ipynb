## Feature scaling

```python
import pandas as pd
```

```python
df = pd.read_csv('for_feature_scaling.csv')
```

```python
df.head(50)
```

```python
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
feature_scaler = [feature for feature in df.columns if feature not in ['Id','SalePrice']]
feature_scaler
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[feature_scaler])
```

```python
df.head()
```

```python
scaler.transform(df[feature_scaler])
```

```python
data = pd.concat([df[['Id','SalePrice']].reset_index(drop=True),
                  pd.DataFrame(scaler.transform(df[feature_scaler]),columns=feature_scaler)],axis=1)
```

```python
data.head()
```

```python
data.to_csv('X_train.csv',index=False)
```
